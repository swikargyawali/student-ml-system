Student Class (Data Representation)

```python
class Student:
    def __init__(self, name, marks, attendance):
        self.name = name
        self.marks = marks
        self.attendance = attendance

## In ML terms this is a data point/sample
```

Dataset Class (Data Handler)

```python
class Dataset:
    def __init__(self,students):
        self.students=students

    def get_all(self):
        return self.students
```

Preprocessing

```python
class Preprocessor:

    def clean(self,students):
        cleaned=[]

        for s in students:
            if 0<=s.marks<=100:
                cleaned.append(s)
        return cleaned
    
    def transform(self,students):
        for s in students:
            s.marks=s.marks/100

        return students
    
## Widely used for feature engineering
```

Model ( Brain of the system )

```python
class Model:
    def train(self,students):
        total=sum([s.marks for s in students])
        self.avg=total/len(students)

    def predict(self,student):
        if student.marks>self.avg:
            return "Pass"
        else:
            return "Fail"
        
## Here Model learns marks and then uses it for prediction
```

Trainer ( Control System )

```python
class Trainer:
    def __init__(self,model):
        self.model=model

    def fit(self,students):
        self.model.train(students)
```

Predictor ( Output Engine )

```python
class Predictor:
    def __init__(self,model):
        self.model=model
        
    def predict_all(self,students):
        results=[]
        for s in students:
            results.append((s.name,self.model.predict(s)))
            return results
        
## Generates the prediction and loop through the data
```

Main Pipeline ( Full Flow )

```python
## Step 1: Creating the data

students = [
    Student("A", 80, 90),
    Student("B", 40, 70),
    Student("C", 60, 85),
    Student("D", 30, 50)
]

## Step 2: Dataset

dataset= Dataset(students)

## Step 3: Preprocessing

pre= Preprocessor()
cleaned= pre.clean(dataset.get_all())
processed=pre.transform(cleaned)

## Step 4: Model
model= Model()

## Step 5: Training
trainer= Trainer(model)

## Step 6: Prediction
predictor= Predictor(model)
results=predictor.predict_all(processed)

## Step 7: Output
for r in results:
    print(r)
```
