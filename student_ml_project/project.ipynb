Student Class (Data Representation)

```python
class Student:
    def __init__(self,name,marks,attendance):
        self.name=name
        self.marks=marks
        self.attendance=attendance


## In ML terms this is a data point/sample
```

Dataset Class (Data Handler)

```python
class Dataset:
    def __init__(self,students):
        self.students=students

    def get_all(self):
        return self.students
```

Preprocessing

```python
class Preprocessor:

    def clean(self,students):
        cleaned=[]

        for s in students:
            if 0<=s.marks<=100:
                cleaned.append(s)
        return cleaned
    
    def transform(self,students):
        for s in students:
            s.marks=s.marks/100

        return students
    
## Widely used for feature engineering
```
